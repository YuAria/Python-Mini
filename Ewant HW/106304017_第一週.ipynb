## 挑戰題一：買相機
### 題目說明：財團法人大兔兔文教基金會要添購攝影設備。
### 他們預計在未來四個月內分別添購四項價格為 100k,150k,230k,80k的設備，
### 請以折線圖畫出累積消費額，並以圓餅圖畫出各器材所佔比例。

```python
camera_prices = [100, 150, 230, 80]
total_camera_prices = [100, 250, 480, 560]
number_of_cameras = [i for i in range(1,5)]
plt.plot(number_of_cameras, total_camera_prices)
plt.xticks(number_of_cameras)
plt.xlabel("Number of equipment")
plt.ylabel("Money spent (in thousands)")
plt.show()
```

```python
plt.pie(x = camera_prices, labels = number_of_cameras, autopct = "%.1f%%")
plt.show()
```

```python
# 我想比較一下中國大陸跟美國2008~2017年每年的穀物生產量(單位: 噸)
year = list(range(2008, 2018))
cereal_usa = [402399936,418666166,401126330,385545256,356210124,434308450,442849090,431866508,475958847,440117262]
cereal_chn = [478546900,481563308,496343192,519373896,539346800,552691792,557417300,621439200,616250500,617930299]
plt.plot(year, cereal_usa, color = "blue")
plt.plot(year, cereal_chn, color = "red")
plt.xticks(year)
plt.fill_between(year, cereal_usa, cereal_chn, color = "red", alpha = 0.2)
plt.fill_between(year, cereal_usa, 3e8, color = "blue", alpha = 0.2)
plt.xlabel("Year")
plt.ylabel("Cereal Production in tons")
plt.show()
# 發現中國大陸的穀物生產量在十年前就已經比美國多了。
```
