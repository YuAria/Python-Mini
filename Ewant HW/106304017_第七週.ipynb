### 我們今天的目標，是利用其中的「性別 Sex」跟「年齡 Age」來預測「存活與否 Survived」(1是存活，0則否）。

1.首先，請使用pandas套件讀檔

2.請取出所有的Age跟Sex作為題目，和Survived做為答案並分成三個同名的numpy array。 (提示： Age.shape = (891, ))

3.其中，由於Age有一些值是空值NaN，請把這些空值都先換成0。提示： 可使用 np.isnan(某個你想確認是不是NaN的年齡) ）

4.再來，Sex部分因為都是文字，不方便我們做svm，請把Male都換成0，Female都換成1 

5.由於我們想把Sex跟Age都當成data，請把Sex和Age兩個array合併成一個。（提示：先用np.reshape，再用np.append）

6.實際拿svm 訓練看看，並用同樣的data，preict看看結果（這次不用分training data跟testing data）

加分題： 請設計一種方法，比較看看你的預測結果，和正確答案(Survived)，找出你的「預測正確率」是多少。

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("train.csv")
```

```python
df
```

```python
sex = np.array(df.Sex)
age = np.array(df.Age)
survived = np.array(df.Survived)
```

```python
for i in range(sex.shape[0]):
    if sex[i] == 'male':
        sex[i] = 0
    else:
        sex[i] = 1
```

```python
sex
```

```python
for i in range(age.shape[0]):
    if np.isnan(age[i]):
        age[i] = 0
```

```python
age
```

```python
age.shape
```

```python
sex = sex.reshape(-1, 1)
age = age.reshape(-1, 1)
sex_age = np.append(sex, age, axis = 1)
```

```python
from sklearn.svm import SVC
```

```python
model = SVC(gamma = 'auto')
model.fit(sex_age, survived)
```

```python
survived_predict = model.predict(sex_age)
model.predict(sex_age)
```

```python
#預測正確率
acc = survived[survived_predict == survived].shape[0]/survived.shape[0]
survived[survived_predict == survived].shape[0]/survived.shape[0]
```

```python
# 也可以直接用model.score
model.score(sex_age, survived)
```

```python
print(f"預測正確率為: {round(acc*100, 2)} %")
```
